### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
archivos_tiendas = [tienda, tienda2, tienda3, tienda4]

ingresos_por_tienda = []

for indice, archivo_tienda in enumerate(archivos_tiendas):
  ingreso_total = archivo_tienda['Precio'].sum()
  ingresos_por_tienda.append(ingreso_total)
  print(f"El ingreso total de la tienda {indice + 1} es: {ingreso_total}")
```

# 2. Ventas por categoría

```python
archivos_tiendas = [tienda, tienda2, tienda3, tienda4]

ventas_por_categoria = []

for indice, archivo_tienda in enumerate(archivos_tiendas):
  ventas_categoria = archivo_tienda.groupby('Categoría del Producto')['Producto'].count()
  ventas_por_categoria.append(ventas_categoria)
  print(f"La cantidad de productos vendidos de la tienda {indice + 1} por categoria son: {ventas_categoria}")
```

```python
numero = tienda.groupby('Categoría del Producto')['Producto'].count()
print(numero)
```

# 3. Calificación promedio de la tienda


```python
archivos_tiendas = [tienda, tienda2, tienda3, tienda4]

calificacion_por_tienda = []

for indice, archivo_tienda in enumerate(archivos_tiendas):
  promedio_calificacion_tienda = archivo_tienda['Calificación'].mean()
  calificacion_por_tienda.append(promedio_calificacion_tienda)
  print(f"La calificación promedio de satisfacción del cliente de la tienda {indice + 1} es: {promedio_calificacion_tienda}")
```

# 4. Productos más y menos vendidos

```python
archivos_tiendas = [tienda, tienda2, tienda3, tienda4]

cantidad_productos = []

for indice, archivo_tienda in enumerate(archivos_tiendas):
  cantidad_productos_tienda = archivo_tienda.groupby(by= 'Producto')['Producto'].count().sort_values(ascending=False)
  producto_mas_vendido = cantidad_productos_tienda.index[0]
  producto_menos_vendido = cantidad_productos_tienda.index[-1]
  print(f"El producto mas vendido de la tienda {indice + 1} es {producto_mas_vendido} y el producto menos vendido de la tienda es {producto_menos_vendido}")
```

# 5. Envío promedio por tienda

```python
archivos_tiendas = [tienda, tienda2, tienda3, tienda4]

costo_de_envio = []

for indice, archivo_tienda in enumerate(archivos_tiendas):
  promedio_costo_envio_tienda = archivo_tienda['Costo de envío'].mean()
  costo_de_envio.append(promedio_costo_envio_tienda)
  print(f"La costo promedio de envio de la tienda {indice + 1} es: {promedio_costo_envio_tienda}")
```
